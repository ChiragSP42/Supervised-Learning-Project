##5

###5 A
Experimenting with various parameters to improve performance.

So the hyperparameters are:
n_neighbors, weights, p, metric, algorithm, leaf_size, n_jobs.

**n_neighbors:** number of neighbors to use.

**weights:** weight function used in prediction.

**p:** power parameter for Minkowski distance.

**metric:** Minkowski distance.

**algorithm:** algorithm used to compute nearest neighbors.

**leaf_size:** leaf size passed to BallTree or KDTree.

**n_jobs:** number of parallel jobs to run for neighbors each.



```python
from sklearn.model_selection import GridSearchCV

k_range = list(range(1, 31))
weight_options = ["uniform", "distance"]
leaf_size = list(range(1,50))
algorithm = ["auto", "ball_tree", "kd_tree", "brute"]
p = [1,2]

param_grid = dict(n_neighbors=k_range, leaf_size=leaf_size, p=p, weights = weight_options)
```

Let us scale the data to increase the efficiency of the model by using MinMaxScaler.

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
KNN_best = KNeighborsClassifier()
```

```python
grid = GridSearchCV(KNN_best, param_grid,cv = 10, scoring = 'accuracy')
grid.fit(X_train_scaled, Y_train)
```

Okay so using GridSearchCV we've computed the best parameters to maximize accuracy.

Let's print these out.

```python
print(f"\x1B[1mBest score:\x1B[0m {grid.best_score_}")
print(f"\x1B[1mBest parameters:\x1B[0m\n {grid.best_params_}")
print(f"\x1B[1mBest estimetor:\x1B[0m\n {grid.best_estimator_}")
```

###5 B
So let's plug these parameters into our KNN to see how much our efficiency has increased.

```python
KNN_best = KNeighborsClassifier(n_neighbors=grid.best_params_['n_neighbors'], leaf_size=grid.best_params_['leaf_size'], p=grid.best_params_['p'], weights=grid.best_params_['weights'])
```

```python
KNN_best.fit(X_train_scaled, Y_train)
```

```python
y_pred_best = KNN_best.predict(X_test_scaled)
```

###3 E
Balancing the output.

There are two ways to achieve this, undersampling and oversampling. The latter is preferred because undersampling may remove some critical data.

There are two ways to oversample the data, either by using SMOTE or ADASYN.

SMOTE:

DISADV:
1. The synthetic instances are generated in the same direction,i.e, connected by an artifical line.

2. Tends to create a large no. of noisy data.

ADASYN:
Generalised form of SMOTE. Difference being it considers the density distribution.

```python
from imblearn.over_sampling import ADASYN

ada = ADASYN(random_state=130)

X_train_ada, Y_train_ada = ada.fit_resample(X_train, Y_train)                                            
```

```python
sns.histplot(Y_train)
```

```python
sns.histplot(Y_train_ada)
```

Using SMOTE and ADASYN, we **increased our recall by 23%** but at the cost of **decreasing precision by 35%.**

So in both models, our False Positives has increased which translates to **more customers being INCORRECTLY identified as HAVING a loan on their card.**

In business it's better to incorrectly miss potential customers than lend out money to customers who we incorrectly identified as potential customers and lose money.
